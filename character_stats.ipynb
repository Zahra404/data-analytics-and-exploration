##### the good characters have the most characters with an intelligence greater than or equals to 100, one of the villains has a higher intelligence than all the other characters 

```python
combat = characters.sort_values('Combat', ascending = False)
combat = combat.iloc[:100]
```

```python
sns.scatterplot(x = 'Intelligence', y = 'Combat', data = combat)
```

##### combat does not correlate with intelligence and some characters have higher combat than characters with high intelligences. A character with the highest combat has an intelligence of  75 

```python
fig = plt.figure(figsize=(10, 8))
sns.scatterplot(x = 'Intelligence', y = 'Combat', data = characters);
```

##### Does the speed of the character have an affect on their combat

```python
speed = characters.sort_values("Speed", ascending = False)
speed = speed.iloc[:100]
```

```python
sns.relplot(x="Speed", y="Combat", kind="line", data=characters)
# used a line graph for variety
```

```python
fig = plt.figure(figsize=(10, 8))
sns.scatterplot(x="Speed", y="Combat", data=speed);
# but scatterplots are easier to read 
```

##### There doesnt seem to be a link between Combat and Speed

##### The characters with an intelligence of 70 seemed to have the most cases of higher speeds

```python
fig = plt.figure(figsize=(10, 8))
sns.distplot(characters.loc[characters['Speed'] >= 60]['Intelligence'], kde_kws={'label': 'Speed >= 60'});
```
